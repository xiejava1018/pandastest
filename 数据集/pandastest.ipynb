```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:yourpassword@localhost:3306/pandastest')
```

```python
sql_query_staff = 'select * from staff;'

df_staff = pd.read_sql_query(sql_query_staff, engine)
df_staff
```

```python
#统计描述
df_staff['staff_salary'].describe()
```

```python
#平均薪水
df_staff['staff_salary'].mean()
```

```python
#df_staff['staff_salary']>10000
```

```python
#月薪过万的人
df_staff[df_staff['staff_salary']<10000]
```

```python
def add_salary(x):
    if x<10000:
        x=x+1000
    return x

df_staff['staff_salary']=df_staff['staff_salary'].map(add_salary)
df_staff
```

```python
sql_query_depatement='select * from depatement;'
df_depatement = pd.read_sql_query(sql_query_depatement, engine)
df_depatement
```

```python
df_staff1=pd.merge(df_staff,df_depatement,left_on='staff_depatment_id',right_on='depatement_id')
df_staff1
```

```python
#按性别统计
df_staff_group=df_staff['staff_id'].groupby(df_staff['staff_gender']).count()
df_staff_group
```

```python
#统计每个部门男女的数目
df_staff_group=df_staff1['staff_id'].groupby([df_staff1['depatement_name'],df_staff1['staff_gender']]).count()
df_staff_group
```
